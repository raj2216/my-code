# Xlsx file handling

- It is a structured data

- csv format file can be created by the Text Editor and Excel because it not having a structured data so it is called as unstructured data
- when use Excel the data will always Structured format because of rows and columns
- xlsx file also called as Workbook

- similar to Book where it contains pages same as workbook contains csv file in xlsx
- where when want to save multiple csv files in single file format that format called as xlsx
- when you are having 1csv and 1 xlsx which you choose we need to choose xlsx because it dont have parser issues and encoding issues and it is pure structured data

# How to convert dataframe into excel

```python
# converting the single dataframe into .xlsx format
# df.to_excel(path)
d[0].to_excel(r"C:\Users\udayk\OneDrive\Documents\dem.xlsx")
```

- converting the multiple dataframe into .xlsx format with different different sheet name
- d[0].to_excel(r"C:\Users\udayk\OneDrive\Documents\dem.xlsx") 
- it is not possible by  to_excel at a time only convert  one df into excel
- when ever we want to save multiple dataframe into multiple sheets
- function is used in pandas to convert is pd.ExcelWriter()  ExcelWriter
- Handling files in python  is handled by in-built func called open() by using which handlig the file
- when we are using open() func this will create a file object this object is always be opened inside the ram
- when we dont close it correctly it will corrupt the data when read write the file and ram consumption is more
- And it mandatory to use f.close()
- to avoid this we have **with** keyword  it automatically closes after the functionality
- Similarlly ExcelWriter() is also a file object where create an excel sheet object that it open/ available inside ram  till you close the sheet for that we use **with** keyword

- f = pd.ExcelWriter(r"C:\Users\udayk\OneDrive\Documents\demmo.xlsx")
- f.close()
- it will be automatically by **with**

```python
with pd.ExcelWriter(r"C:\Users\udayk\OneDrive\Documents\demmo.xlsx") as f:
    d[0].to_excel(f,sheet_name = "Data1")
    d[1].to_excel(f,sheet_name = "Data2")
    d[2].to_excel(f,sheet_name = "Data3")
```

- whenever your writing csv or any xlsx please use mode 
- if dont use mode original data will be corrupted 
- default mode will be "w" 
- use mode = "x" to check  whether  the file is already exists or not


# JSON File handling

### (Java Script Object Notation)

-  JSON is the second largest available file format After csv 
-  When ever we are dealing with APIs, Data Bases , mostly they use file format is JSON , sometimes  it will be xml format
-  JSON is categorized as  Semi - structured data format
-  But JSON can be  --> Structured and  sometimes Semi - structured data
- Mostly it will be in Semi structured form
- JSON  Default Format is same as Dictionary as in python programming
- But there is a small difference in JSON and Dict
- structure of JSON and dict is same it  has keys and values where elemnt is consider as entire and key and value
- main difference in JSON and Dataframe is on keys 
- where always keys should be strings in JSON
- Values can be anything In JSON
- **{"Name" : ["p1","p2"], "Age" : [12,13]}**   it is in JSON format but when excutes in python it will think it is a Dict
- **'{"Name" : ["p1","p2"], "Age" : [12,13]}'** we need to use single quotes in infront of dictionary where we are  using double quotes inside the dict
- Default  JSON format Simplest JSON
- pd.read_json()  Can only convert Structured JSON format to Dataframe
- pd.read_json("path / String of dictionary")
- we cannot convert the semi structured JSON format into Dataframe
- Extension of JSON format file is .json
- 

# {"key" : ["p1","p2"],"key2": [12,13] }
# {"Name" : ["p1","p2"], "Age" : [12,13]}

# Different Formats of JSON

### JSON formats:
1. **'{1 :["p1","p2"], "Age" : [12,13]}'**
2. **{"column": ["c1" ,"c2"], "index" : [0,1], "data" : [[a,b],[c,d]]}**
3. **{"name" : {"o" : p1 , "1" : p2} , "Age" : {"0" : 12, "1" : 14}}**
4. **{"0" : {"name" : p1 , "Age" : 12}, "1" : {"name" : p2 , "Age" : 14}}**
5. **[[],[],[]]**

```python
# Most occuring format of semi - structured data
z2 = {"Name" : "p1","Marks": [{"Sem1" : 20, "Sem2" : 24}], "Age" : 58}
```

```python
# record path is only used when the values are in list of dict
pd.json_normalize(z2,record_path = ["Marks"] )
```

```python
# parllel we need to use meta for remaing columns
pd.json_normalize(z2,record_path = ["Marks"], meta = ["Name","Age"])  
```
